## About NETFLIX

Netflix is one of the most popular media and video streaming platforms. They have over 10000 movies or tv shows available on their platform, as of mid-2021, they have over 222M Subscribers globally. This tabular dataset consists of listings of all the movies and tv shows available on Netflix, along with details such as - cast, directors, ratings, release year, duration, etc.

### Business Problem

Analyze the data and generate insights that could help Netflix in deciding which type of shows/movies to produce and how they can grow the business in different countries.

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
```

```python
df['date']=df['date_added'].dt.day
df['month']=df['date_added'].dt.month
df['year']=df['date_added'].dt.year
```

```python
# Content Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', hue='type', palette='coolwarm', legend=False)
plt.title("Distribution of Movies vs TV Shows")
plt.show()
```

```python
# Genre Analysis
genre_counts = df['listed_in'].str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=genre_counts.values, y=genre_counts.index, hue=genre_counts.index, palette='viridis', legend=False)
plt.title("Top 10 Most Common Genres")
plt.xlabel("Count")
plt.show()
```

```python
# Release Trend Analysis
plt.figure(figsize=(12, 5))
sns.histplot(df['release_year'], bins=30, kde=True, color='skyblue')
plt.title("Content Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# Country-wise Content Distribution
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, hue=country_counts.index, palette='magma', legend=False)
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Count")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['Duration_numeric'] = df['duration'].str.extract('(\d+)')[0].astype(float)
sns.boxplot(data=df, x="type", y="Duration_numeric", hue="type", palette="Set2", legend=False)
plt.title("Duration Analysis of Movies vs TV Shows")
plt.ylabel("Duration (Minutes or Seasons)")
plt.show()
```

### Key Insights:
1. Netflix has been increasing its focus on TV Shows over time.
2. The most common genres include Drama, Comedy, and Action.
3. The USA, India, and UK produce the highest number of Netflix content.
4. Content releases peaked in recent years, indicating aggressive expansion.
5. The correlation heatmap suggests no strong numerical relationships, but categorical trends are valuable.

### Business Recommendations:
1. Invest more in trending genres like Drama and Comedy, as they dominate the platform.
2. Expand localized content in emerging markets like India, as it has a strong production presence.
3. Leverage data on seasonal trends to strategically time content releases for maximum engagement.
4. Focus on producing TV shows, as they have seen increasing demand over time.
5. Partner with popular directors and actors in top-producing countries to boost regional viewership.
